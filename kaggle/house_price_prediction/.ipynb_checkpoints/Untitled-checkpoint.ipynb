```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from datetime import datetime
import seaborn as sns
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
```

```python
train_data = pd.read_csv('train.csv')
test_data = pd.read_csv('test.csv')
train_data.head()
```

```python
test_data.head()
```

```python
train_data.info()
```

```python
missing_fraction = train_data.isna()
missing_fraction = missing_fraction.sum()/len(train_data)
missing_fraction[missing_fraction > 0.0].sort_values(ascending=False)
```

```python
# we will remove all features where there are more than 40% missing data points, since filling with means or mode 
# will not yield good values
missing_fraction_dict = dict(missing_fraction[missing_fraction > 0.4].sort_values(ascending=False))
train_data = train_data.drop(missing_fraction_dict.keys(), axis=1)
train_data = train_data.drop(['Id'], axis=1)
train_data.Condition2.value_counts()
```

```python
#Generate Pearson correlation matrix for visualization
plt.figure(figsize=(12,10))
cor = train_data.corr().round(3)
sns.heatmap(cor, cmap=plt.cm.Reds)
plt.show()

cor_target = abs(cor["SalePrice"])
#Selecting highly correlated features
relevant_features_corr = cor_target[abs(cor_target)>0.5] # relevant features as per the correlation matrix
print(relevant_features_corr.sort_values(ascending=False))
relevant_features_dict = dict(relevant_features_corr.sort_values(ascending=False))
```

```python
from math import ceil as ceil
numerical_features = list(cor)
numerical_features.remove('SalePrice') # remove the dependent variable
num_features = len(numerical_features)
numerical_features_std = dict.fromkeys(numerical_features)
ncols=2
nrows = ceil(num_features/ncols)

    
fig, axes = plt.subplots(ncols=ncols, nrows=nrows, figsize=(12,40))
                
for ax,feature in zip(axes.flatten(),numerical_features):
    plt.sca(ax)
    sns.distplot(train_data[feature])
    ax.set_title(feature)
    std = np.round(np.std(train_data[feature]), 2)
    numerical_features_std[feature] = std
    ax.legend(['std: {}'.format(std)])

fig.tight_layout(pad=3.0)
```

```python
# based on the fact that several features have very strong outliers we will use median to fill NaN
imputer = SimpleImputer(missing_values=np.nan, strategy='median')
imputer = imputer.fit(train_data[numerical_features])
train_data[numerical_features] = imputer.transform(train_data[numerical_features])
test_data[numerical_features] = imputer.transform(test_data[numerical_features])

#scale features to mean 0 and std 1
scaler = MinMaxScaler()
train_data[numerical_features] = scaler.fit_transform(train_data[numerical_features])  
test_data[numerical_features] = scaler.transform(test_data[numerical_features])

# plot again to confirm it worked
fig, axes = plt.subplots(ncols=ncols, nrows=nrows, figsize=(12,40))
for ax,feature in zip(axes.flatten(),numerical_features):
    plt.sca(ax)
    sns.distplot(train_data[feature])
    ax.set_title(feature)
    std = np.round(np.std(train_data[feature]), 2)
    numerical_features_std[feature] = std
    ax.legend(['std: {}'.format(std)])

fig.tight_layout(pad=3.0)
```
